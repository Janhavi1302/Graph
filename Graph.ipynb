```python
import matplotlib.pyplot as plt
```

```python
Climate=['rainy','sunny','humid','dry','cloudy']
color=['blue','red','green','yellow','black']
slice=[3,4,5,2,1]
plt.pie(slice ,colors=color,labels =Climate, startangle = 90 , shadow = True , explode = (0,0.1,0,0,0) , radius = 1, autopct = '%1.0f%%') 
plt.legend()
plt.show()
```

```python
yaxis=[3,9,7,6,1]
xaxis=[1,2,3,4,5]
plt.plot(xaxis,yaxis)
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('single line graph')
plt.show()

```

```python
x2=[1,2,3,4,5]
y2=[3,5,6,7,1]
plt.plot(x2, y2, label = "Janhavi")
x1=[2,4,5,6,5]
y1=[8,7,6,3,1]
plt.plot(x1,y1, label = "Ram")
plt.legend()
plt.title('double line graph')
plt.show()
```

```python
yaxis=[3,9,7,6,2]
xaxis=[8,2,3,4,5]
plt.plot(xaxis,yaxis,color = 'pink',linestyle = "dashed",linewidth = 2,marker = 'o',markerfacecolor = 'black',markersize = 10)
plt.ylim(1,10)
plt.xlim(1,10)
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('dotted line graph')
plt.show()
```

```python
age = [10,20,30,40,50]            #xaxis
child = [2,5,8,15,20]          #yaxis
label = ['one','two','three','four','five']
plt.bar(age ,child ,label = label ,width = 2 ,color = ['yellow','orange'])
plt.xlabel('age')
plt.ylabel('child')
plt.title('My bar graph')
plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [1, 5, 3, 6, 7, 13, 14, 15, 18, 19, 25, 28, 30, 21, 39, 36, 35, 40, 45, 48, 50]
range_vals = (0, 50)
bins = 10

# Create histogram
plt.hist(marks, bins=bins, range=range_vals, color='purple', histtype='bar', rwidth=0.7)

# Add labels and title
plt.xlabel('Marks')
plt.ylabel('No of Students')
plt.title('Histogram')

# Show plot
plt.show()
```
